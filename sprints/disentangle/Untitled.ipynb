```python
import numpy as np
import sklearn.datasets
import sklearn.preprocessing
import sklearn.decomposition
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
```

```python
ica = sklearn.decomposition.FastICA(n_components=10, whiten=True, max_iter=500000)
D = ica.fit_transform(np.reshape(x_train, [60000, -1]))
```

```python
plt.figure(figsize=(16,4))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(np.reshape(ica.components_[i], [28,28]), cmap='gray', interpolation='nearest')
    plt.axis('off')
```
