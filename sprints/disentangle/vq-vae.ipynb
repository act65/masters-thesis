```python
import tensorflow as tf
import sonnet as snt
tf.enable_eager_execution()
```

```python
B = 50
D = 8
n = 32
bits = 4
x = tf.random_normal([B, D, n])
```

```python
vqvae = snt.nets.VectorQuantizer(commitment_cost=1.0, num_embeddings=bits, embedding_dim=D)
```

```python
vqvae(tf.unstack(x, axis=-1)[0], True).keys()
```

```python
code = tf.stack([tf.argmax(vqvae(z, True)['encodings'], axis=-1) for z in tf.unstack(x, axis=-1)], axis=-1)
y = tf.stack([vqvae(z, True)['quantize'] for z in tf.unstack(x, axis=-1)], axis=-1)
```
